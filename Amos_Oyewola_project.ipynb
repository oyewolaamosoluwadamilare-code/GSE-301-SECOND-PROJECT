# 1. Variables & File Handling

```python

filename = "data.csv"  # can be .txt or .csv

data = []

try:
    with open(filename, "r") as file:
        for line in file:
            line = line.strip()
            if line != "":
                data.append(float(line))

    print("Data loaded successfully:")
    print(data)

except FileNotFoundError:
    print("Error: File does not exist.")
```

# 2. Error Handling

```python

filename = "data.csv"
data = []

try:
    with open(filename, "r") as file:
        for line in file:
            line = line.strip()
            if line != "":
                try:
                    data.append(float(line))
                except ValueError:
                    print("Warning: Invalid (non-numeric) value skipped.")

    if len(data) == 0:
        print("Error: File is empty or contains no valid numeric data.")
    else:
        print("Valid data:", data)

except FileNotFoundError:
    print("Error: File not found.")
```

# 3. Functions

```python
def calculate_total(data):
    total = 0
    for value in data:
        total += value
    return total

def calculate_average(data):
    total = calculate_total(data)
    return total / len(data)

def calculate_minimum(data):
    minimum = data[0]
    for value in data:
        if value < minimum:
            minimum = value
    return minimum

def calculate_maximum(data):
    maximum = data[0]
    for value in data:
        if value > maximum:
            maximum = value
    return maximum


# Testing the functions
data = [45, 67, 89, 34, 72]

print("Total:", calculate_total(data))
print("Average:", calculate_average(data))
print("Minimum:", calculate_minimum(data))
print("Maximum:", calculate_maximum(data))
```

# Question 4: Operators & Loops

```python
data = [45, 67, 89, 34, 72]

total = 0
count = 0
minimum = data[0]
maximum = data[0]

for value in data:
    total += value
    count += 1

    if value < minimum:
        minimum = value
    if value > maximum:
        maximum = value

average = total / count

print("Total:", total)
print("Average:", average)
print("Minimum:", minimum)
print("Maximum:", maximum)
```

# Question 6: Sets

```python

filename = "categories.txt"
categories = set()

try:
    with open(filename, "r") as file:
        for line in file:
            line = line.strip()
            if line != "":
                categories.add(line)

    print("Unique categories:", categories)
    print("Total number of unique categories:", len(categories))

except FileNotFoundError:
    print("Error: File not found.")
```

# Question 7: Object-Oriented Programming (OOP)

```python
class DataSet:
    def __init__(self, filename):
        self.filename = filename
        self.data = []
        self.total = None
        self.average = None
        self.minimum = None
        self.maximum = None

    def load_data(self):
        try:
            with open(self.filename, "r") as file:
                for line in file:
                    line = line.strip()
                    if line != "":
                        self.data.append(float(line))
        except FileNotFoundError:
            print("Error: File not found.")

    def calculate_statistics(self):
        if len(self.data) == 0:
            print("No data to calculate.")
            return

        self.total = sum(self.data)
        self.average = self.total / len(self.data)
        self.minimum = min(self.data)
        self.maximum = max(self.data)

    def display_results(self):
        if self.total is None:
            return

        print("Total:", self.total)
        print("Average:", self.average)
        print("Minimum:", self.minimum)
        print("Maximum:", self.maximum)


dataset = DataSet("data.csv")
dataset.load_data()
dataset.calculate_statistics()
dataset.display_results()
```
